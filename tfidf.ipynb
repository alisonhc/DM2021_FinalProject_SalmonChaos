```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def rank_most_similar_documents(input_doc_embedding, all_docs_embeddings):
    """
    Get top n matched documents plus their indexes
    Return list of tuples where first position in each tuple is index and second position is the probability
    """
    distances = cosine_similarity(input_doc_embedding, all_docs_embeddings)[0]
    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1], reverse=True)
    res_with_doc_index = []
    return results
```

### Try to figure out what probability threshold to use based on train labels

```python
train_result_df = pd.DataFrame()
```

```python
def rank_all_embeddings(all_embeddings, doc_index_list):
    rank_probs_all = []
    rank_inds_all = []
    for i in range(len(doc_index_list)):
        rank_results = rank_most_similar_documents(all_embeddings[i], all_embeddings)
        rank_probs = []
        rank_inds = []
        for ind,prob in rank_results:
            if ind != i:
                rank_probs.append(prob)
                rank_inds.append(doc_index_list[ind])
        rank_probs_all.append(rank_probs)
        rank_inds_all.append(rank_inds)
    return rank_probs_all, rank_inds_all
```

```python
rank_probs_all, rank_inds_all = rank_all_embeddings(tfidf_embeddings, prep_df['doc_index'])
train_result_df['doc_index'] = prep_df['doc_index']
train_result_df['rank_inds'] = rank_inds_all
train_result_df['rank_probs'] = rank_probs_all
train_result_df['actual_result'] = sim_doc_col
```

### Load test data and create result file with prob threshold

```python
# read in test data
test_df = pd.read_csv("public_test_data_processed.csv")
test_df[0:5]
```

```python
THRESH = 0.8
```

```python
test_tokens = test_df['tokens_rep']
test_embeddings = tfidf.transform(test_tokens)
```

```python
test_rank_probs, test_rank_inds = rank_all_embeddings(test_embeddings, test_df['doc_index'])
```

```python
probs_to_save = []
inds_to_save = []
for i in range(len(test_rank_inds)):
    this_ind_probs_to_save = []
    this_ind_inds_to_save = []
    this_ind = 0
    prob = test_rank_probs[i][this_ind]
    while prob >= THRESH:
        this_ind_probs_to_save.append(prob)
        this_ind_inds_to_save.append(test_rank_inds[i][this_ind])
        this_ind += 1
        prob = test_rank_probs[i][this_ind]
    probs_to_save.append(this_ind_probs_to_save)
    inds_to_save.append(this_ind_inds_to_save)
```
