```python
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("processed_data.csv")
df.head()
```

```python
def get_county(raw_text):
    county = { 1: ['臺北', '台北'], 2: ['新北'], 3: ['桃園'],  4: ['臺中', '台中'],5: ['臺南', '台南'],6: ['高雄', '高屏'], 
            7: ['新竹'],  8: ['苗栗'], 9: ['彰化'], 10: ['南投'], 11: ['雲林'], 12: ['嘉義'], 13: ['屏東', '高屏'], 14: ['宜蘭'], 
            15: ['花蓮'], 16: ['臺東', '台東'],17: ['澎湖'], 18: ['金門'], 19: ['連江'], 20: ['基隆'],  21: ['嘉義']}
    ret = np.zeros(22, dtype=int)
    for class_ in county:
        for seed in county[class_]:
            if seed in raw_text:
                ret[class_-1] = 1
    return ret
```

```python
tmp = df['raw_text'][1]
get_county(tmp)
```

```python
d_len = len(df)
df["county"] = ""
for i in range(d_len):
    df["county"][i] = get_county(df['raw_text'][i])
```
